```python
import pandas as pd
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
ttestdf = pd.read_excel('wine_data_all.xlsx')
ttestdf.head()
```

```python
primeurstrue_df = ttestdf.loc[ttestdf['is_primeurs'] == True]
primeurstrue_df.head()
primeurstrue_mean = primeurstrue_df.score.mean()
print(primeurstrue_mean)
```

```python
primeursfalse_df = ttestdf.loc[ttestdf['is_primeurs'] == False]
primeursfalse_df.head()
primeursfalse_mean = primeursfalse_df.score.mean()
print(primeursfalse_mean)
```

```python
stats.ttest_ind(primeurstrue_df.score, primeursfalse_df.score, equal_var=False)
# Since our p-value is less than 0.05, our Null Hypothese is false (reject that sh**). There is statistical differencce.
```
