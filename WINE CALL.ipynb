```python
# Dependencies
import requests
import pandas as pd
from config import api_key
```

```python
# Build query URL and request your results in Celsius
url = 'https://api.globalwinescore.com/globalwinescores/latest/?'
url2 = '&limit=10000&wine_id>999'

query_url = f'{url}{url2}'
# Get weather data
response = requests.get(query_url, headers={'Authorization' : f'Token {api_key}'})
```
